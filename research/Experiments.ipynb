```python
import os
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python

data_dir = "../research/V1/images/train"
image_files = [f for f in os.listdir(data_dir) if f.endswith((".jpg", ".png"))]
print("Found images:", image_files)
```

```python
num_images = min(1, len(image_files))

plt.figure(figsize=(12, 6))

for i in range(num_images):
    img_path = os.path.join(data_dir, image_files[i])
    img = mpimg.imread(img_path)
    
    plt.subplot(2, 3, i+1)  # 2 rows, 3 columns
    plt.imshow(img)
    plt.axis('off')          # hide axes
    plt.title(image_files[i])

plt.tight_layout()
plt.show()
```

```python
from ultralytics import YOLO
import mlflow
import mlflow.pytorch
```

```python
mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment("yolo_nano_v1_experiment")

with mlflow.start_run():
    
    
    model = YOLO("yolov8n.pt")  # pretrained weights
    
    model.train(
        data={
            'train': '../research/V1/images/train/images',
            'val': '../research/V1/images/val/images'
        },
        imgsz=320,
        batch=1,
        epochs=1,
        name="yolo_nano_v1"
    )
    
    mlflow.pytorch.log_model(model.model, "yolo_nano_model")
    
    # Log parameters
    mlflow.log_param("imgsz", 320)
    mlflow.log_param("batch_size", 1)
    mlflow.log_param("epochs", 1)
```
